```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

# 1.Categorical data


## a.Countplot

```python
sns.countplot(x='Survived' ,data=df)
plt.grid(True)
df['Survived'].value_counts()
```

```python
sns.countplot(x='Pclass', data=df,palette='pastel')
```

## b. Pichart

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Pclass'].value_counts().plot(kind='pie' , autopct='%.2f')
```

# 2.Numerical data

## a.Histogram

```python
plt.hist(df['Age'], bins=5, color='skyblue', edgecolor='black', linewidth=1.2)
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## b.Distplot

```python
sns.distplot(df['Age'])
```

```python
sns.histplot(df['Age'], bins=20, kde=True, edgecolor='black')
plt.title('Distribution of Passenger Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(df['Age'], bins=20, stat='density', color='skyblue', edgecolor='black')
sns.kdeplot(df['Age'], color='black', linewidth=2)
plt.title('Age Distribution with Black KDE Curve')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

## c.Boxplot

```python
sns.boxplot(x='Fare' , data=df)
```
